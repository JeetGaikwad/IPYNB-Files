```python
#NaN - Not a Number (Null value)
#But in a few datasets , NaN is considered as missing value
#When NaN comes into picture,Int is converted into float
```

```python
import numpy as np
df = pd.DataFrame({'Cricket':[1,2,np.nan,4,6,7,2,np.nan],
                   'BaseBall':[5,np.nan,np.nan,5,7,2,4,5],
                   'Tennis':[1,2,3,4,5,6,7,8]})
df
```

```python
#Now let us drop all the null values
df.dropna()
```

```python
#Now let us replace all the null values with 0
df = df.fillna(value = 0)
```
