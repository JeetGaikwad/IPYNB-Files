```python
#LOSS SHOULD REACH MINIMUM AND AT MINIMUM LOSS , WE GET THE BEST FIT LINE
from sklearn.datasets import make_regression
import numpy as np
```

```python
#GRADIENT DESCENT TELLS US THAT NO MATTER WHERE YOU START ,IF YOU TAKE CORRECT STEPS,YOU END UP IN A REQUIRED POSITION.
X,y = make_regression(n_samples = 4,n_features = 1,n_informative = 1,n_targets = 1,noise = 80,random_state = 13)#directly taking i/p and o/p
#n_samples - The Number of Samples
#n_feature - The number of feature(inputs)
#n_informative -Number of informative features,i.e,the number of features used to build the linear model
#n_targets - The number of Regression targets(output)
#noise - The Standard deviation of the Gaussian Noise applied to the output
#random_state - It determines random number generation for dataset creation
```

```python
#VISUALISATION
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
X
```

```python
y
```

```python
#Lets apply Ordinary Least Squared method
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Model fitting
model.fit(X,y)
```

```python
#Let us predict the output
model.predict(X)
```

```python
y
```

```python
#Let me just find out slope m
model.coef_ #slope(m)
```

```python
#y-intercept(C)
model.intercept_
```

```python
#BEST FIT LINE
plt.scatter(X,y)
plt.plot(X,model.predict(X),color = 'red')
```

```python
#LETS APPLY GRADIENT DESCENT ,SSUMING SLOPE IS CONTANT at M = 78.35
#Lets assume the starting value for intercept c = 0
#Taking intercept from a random value(at 0),at the end we will reach near 26.15
y_pred = ((78.35 * X) + 0)
print(y_pred)
#Now I want to reshape the output
y_pred = ((78.35 * X) + 0).reshape(4)
print(y_pred) # we convert it  into vector
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color = 'red',label = 'OLS = 26.15') #BEST FIT LINE
plt.plot(X,y_pred,color = 'lime',label = 'C = 0') #GRADIENT DESCENT AT C = 0
plt.show()
plt.legend()
```

```python
#Iteration1/Epoch 1
#By changing the value of C ,my line will move up or down
#By chaning the value of m , my line will rotate(the slope is varied)
m = 78.35 #SLOPE FIXED
c = 0 #THIS WILL CHANGE
#loss slope = -2 *np.sum(y-m*X.ravel() - c)
loss_slope = -2*np.sum(y-m*X.ravel() - c)
loss_slope
```

```python
#Lets take learning rate as 0.1
lr = 0.1
#Step size = learning rate * slope
step_size = lr* loss_slope
step_size
```

```python
#Now Once we have got the step size , let us calculate the new intercept
c = c - step_size
c
```

```python
#Let us plot the updated intercept value
y_pred1 = ((78.35 * X) + c).reshape(4)
y_pred1
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color = 'red',label = 'OLS = 26.15')
plt.plot(X,y_pred,color = 'lime',label = 'C = 0')
plt.plot(X,y_pred1,color = 'Orange',label = 'C = 20.92')
plt.legend()
plt.show()
```

```python
#Iteration2/Epoch 2
loss_slope = -2 *np.sum(y -m*X.ravel() - c)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
c = c - step_size
c #The send updated intercept value
```

```python
y_pred2 = ((78.35 * X) + c).reshape(4)
y_pred2
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color = 'red',label = 'OLS = 26.15')
plt.plot(X,y_pred,color = 'lime',label = 'C = 0')
plt.plot(X,y_pred1,color = 'Orange',label = 'C = 20.92')
plt.plot(X,y_pred2,color = 'k',label = 'C = 25.11')
plt.legend()
plt.show()
```
