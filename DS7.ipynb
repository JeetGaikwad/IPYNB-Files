```python
#IRIS FLOWER DATASET - CLASSIFICATION - LOGISTIC REGRESSION
#dataset - IRIS FLOWER DATASET
#dataset - https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/IRIS.csv
```

```python
#1.Take the data and create  dataframe
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/IRIS.csv')
df
```

```python
#2.Step no 2 is not required here
#STEP NO 2 is optional
```

```python
#3.Step no 3 is also not required here
#We cannot apply visualisation for CLASSIFICATION MODEL.
```

```python
#Input - Sepal length,Sepal width,petal length and petal width.
#Output - Species
```

```python
#4.Divide the data into input and output
x = df.iloc[:,0:4].values
y = df.iloc[:,4].values
```

```python
#5.train and test variables
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state =0)
#random_state - is used to avoid duplicates
```

```python
#6.NORMALISATION or SCALING
#Here step no 6 is nor required because our inputs are already scaled
```

```python
#7.Apply A CLASSIFIER,regressor or clusterer
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#8.fit the model
model.fit(x_train,y_train)
```

```python
#9.predict the output
y_pred = model.predict(x_test) #using the input testing values , we predict the output.
y_pred #PREDICTED OUTPUT VALUES
```

```python
y_test #ACTUAL OUTPUT VALUES
```

```python
#10.Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
