```python
%%writefile app.py
import streamlit as st
import joblib
model = joblib.load('spam-ham')#We are loading the pipelined model using joblib
st.title('SPAM-HAM CLASSIFIER')#creates a title
ip = st.text_input('Enter the message')#creates a text box in the webapp
op = model.predict([ip])
if st.button('Predict'):
  st.title(op[0])
```
